```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
import os

plt.rcParams.update({'font.size': 22})
plt.rcParams.update({'text.usetex': True})
os.makedirs("plots", exist_ok=True)

# Load all magnetization files
files = glob.glob("results/magnetization_*.txt")

lambda_val = 0.
c_val = 2

plt.figure(figsize=(10, 6.25))
for file in files:
    data = pd.read_csv(file, sep=" ", header=None, names=["Time", "Magnetization"])
    plt.plot(data["Time"], data["Magnetization"])

# Customize plot
# plt.title("Magnetization Over Time for Multiple Runs", fontsize=16)
plt.xlabel(r"$t$")
plt.ylabel(r"$m$")
plt.grid(True, linestyle="--", alpha=0.6)
plt.plot([], [], ' ', label=fr"$\lambda = {lambda_val:.3g}$"+"\n"+fr"$c = {c_val}$")
leg = plt.legend(loc='upper right', handlelength=0, handletextpad=0, fancybox=True)
for item in leg.legend_handles:
    item.set_visible(False)
plt.tight_layout()
# plt.xscale("log")
# Show plot
plt.gca().set_ybound(-1., 1.)
plt.savefig(f"plots/vmpi_phenom_lambda={lambda_val:.3g}.pdf", 
            transparent=True, bbox_inches="tight")
```
